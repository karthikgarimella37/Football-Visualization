```python
fig,ax = plt.subplots()

ax.plot( df.Squad, df.xG, color="red", marker="o")

ax.set_xlabel("Time Period",fontsize=14)

ax.set_ylabel("xG",color="red",fontsize=14)

ax2=ax.twinx()

ax2.plot(df.Squad, df.Gls, color="blue",marker="o")
ax2.set_ylabel("Goals",color="blue",fontsize=14)
plt.show()

fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```

```python
sns.jointplot(x='Gls',y='xG',data=df,kind='reg',scatter_kws={'s':50})
for x,y in zip(df.Gls,df.xG):

    label = f"({x},{y})"

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,5), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
```
