```python
#colour value based on GD
colorsValue_17_18 = []
colorsValue_18_19 = []
colorsValue_19_20 = []
colorsValue_20_21 = []
colorsValue_21_22 = []
colorsValue_22_23 = []

for value in df_17_18.GD:
    if value > 0:
        colorsValue_17_18.append('green')
    elif value == 0.0:
        colorsValue_17_18.append('grey')
    else:
        colorsValue_17_18.append('red')

for value in df_18_19.GD:
    if value > 0:
        colorsValue_18_19.append('green')
    elif value == 0.0:
        colorsValue_18_19.append('grey')
    else:
        colorsValue_18_19.append('red')

for value in df_19_20.GD:
    if value > 0:
        colorsValue_19_20.append('green')
    elif value == 0.0:
        colorsValue_19_20.append('grey')
    else:
        colorsValue_19_20.append('red')
        

for value in df_20_21.GD:
    if value > 0:
        colorsValue_20_21.append('green')
    elif value == 0.0:
        colorsValue_20_21.append('grey')
    else:
        colorsValue_20_21.append('red')
        

for value in df_21_22.GD:
    if value > 0:
        colorsValue_21_22.append('green')
    elif value == 0.0:
        colorsValue_21_22.append('grey')
    else:
        colorsValue_21_22.append('red')
        

for value in df_22_23.GD:
    if value > 0:
        colorsValue_22_23.append('green')
    elif value == 0.0:
        colorsValue_22_23.append('grey')
    else:
        colorsValue_22_23.append('red')
```

```python
#figure size
ax = plt.figure(figsize = (15, 5), dpi = 100.0, facecolor='white', edgecolor=None)
#plot theme
plt.style.use('fivethirtyeight')
#bar plot
ax1 = plt.bar(df_17_18.Matchday, df_17_18.GD, color = colorsValue_17_18, width = 1)



#removing default grid
plt.grid(False)
#removing y axis
plt.xticks([])

#customized legends
W = mlines.Line2D([], [], color='green', marker='s', ls='', label='Win')
D = mlines.Line2D([], [], color='grey', marker='s', ls='', label='Draw')
L = mlines.Line2D([], [], color='red', marker='s', ls='', label='Loss')
#plotting the legends
plt.legend(handles=[W, D, L], loc= 'upper right')

plt.ylabel('Results')
plt.xlabel('Matchday')

```
