```python
df2 = df2[df2.Total != 0]
```

```python
df2 = df2.sort_values(by=['Total'], ascending=False)
df2
```

```python
data3=dict(zip(df2['Player'], df2['Total']))
```

```python
type(data3)
```

```python
K=1
data2=dict()
for key in data3:
      
    # rounding to K using round()
    data2[key] = round(data3[key], K)
```

```python
data2
```

```python
# waffle chart
fig = plt.figure(FigureClass=Waffle,
                 rows=4,
                 #columns=20,
                 values=data2  ,
                 title={
        'label': 'Man United Women xG and xA Contribution\n',
        'loc': 'center',
        'fontdict': {
            'fontsize': 100
        }
    },
          legend={
        'labels': [f"{k} \n({v}) xG+xA" for k, v in data2.items()],  # lebels could also be under legend instead
        'loc':'lower left',
        'bbox_to_anchor': (0, -0.3),
        'ncol': len(data2),
        'framealpha': 2,
        'fontsize': 50
    }    ,
            cmap_name=('tab10'),
                figsize=(150,150))
```
