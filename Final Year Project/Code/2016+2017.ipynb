```python
df_xG=df.xG
df_xGA=df.xGA
df_xG_Diff=df_xG-df_xGA

print((df_xG_Diff).to_string(index=False),"\n\n",(df.WDL).to_string(index=False))
```

```python
pos_count=0
neg_count =0
null_count=0
for i in df_xG_Diff:
    if i > 0: 
        pos_count += 1
    elif i < 0: 
        neg_count += 1
    else:
        null_count+=1
print("Positive xG in 38 matches: ",pos_count)
print("Negative xG in 38 matches: ",neg_count)
print("Null Difference in xG in 38 matches: ",null_count)
```

```python
df_FTWG=df.FTWG
df_FTLG=df.FTLG
df_GD=df_FTWG-df_FTLG
print((df_GD).to_string(index=False),"\n\n\n",df.WDL)
```

```python
one_count=0
two_count=0
three_count=0
for i in df_GD:
    if i==1:
        one_count+=1
    elif i==2:
        two_count+=1
    elif i==3:
        three_count+=1
        
print("GD 1 : ",one_count)
print("GD 2 : ",two_count)
print("GD 3 : ",three_count)
```

```python
pos_count=0
neg_count =0
null_count=0
for i in df_GD:
    if i > 0: 
        pos_count += 1
    elif i < 0: 
        neg_count += 1
    else:
        null_count+=1
print("Positive GD in 38 matches i.e. Wins: ",pos_count)
print("Negative GD in 38 matches i.e. Loss: ",neg_count)
print("Null Difference in GD in 38 matches i.e. Draws: ",null_count)
```

-ve PPDA means they pressed better and with intensity

+ve PPDA means they pressed worse and with lesser intensity

```python
df_PPDA=df.PPDA
df_aPPDA=df.aPPDA
df_PPDA_Diff=df_PPDA-df_aPPDA
print((df_PPDA_Diff).to_string(index=False),df.WDL)
```

```python
pos_count=0
neg_count =0
null_count=0
for i in df_PPDA_Diff:
    if i > 0: 
        pos_count += 1
    elif i < 0: 
        neg_count += 1
    else:
        null_count+=1
print("Positive PPDA in 38 matches : ",pos_count)
print("Negative PPDA in 38 matches : ",neg_count)
print("Null Difference in PPDA in 38 matches : ",null_count)
```

```python
labels=np.array(['FTWG','FTLG','xG','xGA'])
stats=df.loc[15,labels].values
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
# close the plot
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))
```

```python
fig=plt.figure()
ax = fig.add_subplot(111, polar=True)  
ax.plot(angles, stats, 'o-', linewidth=2)  
ax.fill(angles, stats, alpha=0.25) 
ax.set_thetagrids(angles * 180/np.pi, labels) 
ax.set_title([df.loc[15,"HomeTeam"]]) 
#ax.set_rlim(0,250)
ax.grid(True)

#For 36th Match of Leicester City
```
